# Pivot Table

피어슨 상관분석을 위해 병합한 데이터셋을 피벗테이블로 전환

```python
matrix = data.pivot_table(index='Author', columns='Title', values='Rating')
matrix.head(20)
```

피어슨 상관관계 분석

```python
# Pearson Corelation

#                         s1           s2
# Author / Title        영화1        영화2
# ---------------------------------------------
#   a                    2.5
#   b                    5.0          4.5
#   c                    4.5          1.5
#   d
# ---------------------------------------------
#  mean                  3.0          3.0
#                      s1.mean()   s2.mean()



#          s1_c = s1 - s1.mean()   s2_c = s2 - s2.mean()

# Author / Title        s1_c          s2_c          s1_c * s2_c
# -----------------------------------------------------------------
#   a                   -0.5
#   b                    2.0          1.5           3.0 (양의 상관관계)
#   c                    1.5         -1.5          -2.25 (음의 상관관계)
#   d
# -----------------------------------------------------------------
#  sum                                              0.75
```

```python
GENRE_WEIGHT = 0.1

def pearsonR(s1, s2):
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))
    # 분모가 클수록 상관관계가 떨어지고, 분자가 클수록 상관관계가 높아짐

# 추천함수 생성
def recommend(input_movie, matrix, n, similar_genre=True):
    input_genres = data[data['Title'] == input_movie]['Genre'].iloc(0)[0]

    result = []
    for title in matrix.columns:
        if title == input_movie:
            # 같은 영화를 추천받는 경우는 건너뜀
            continue

        # rating comparison
        cor = pearsonR(matrix[input_movie], matrix[title])
        
        # genre comparison
        if similar_genre and len(input_genres) > 0:
            temp_genres = data[data['Title'] == title]['Genre'].iloc(0)[0]
            # 현재 비교하고 있는 영화
            same_count = np.sum(np.isin(input_genres, temp_genres))
            # Input으로 넣은 영화와 비교
            # np.isin() -> 배열을 비교하여 똑같은 요소가 있으면 True를 반환
            cor += (GENRE_WEIGHT * same_count)
        
        if np.isnan(cor):
            continue
        else:
            result.append((title, '{:.2f}'.format(cor), temp_genres))
            
    result.sort(key=lambda r: r[1], reverse=True)
    # 리스트 정렬

    return result[:n]
```

```python
recommend_result = recommend('블랙 위도우', matrix, 10, similar_genre=True)
# '랑종'를 보고 추천영화 10개를 선정
pd.DataFrame(recommend_result, columns = ['Title', 'Correlation', 'Genre'])
```

```python
rr = pd.DataFrame(recommend_result)
rr = rr.loc[0:9, 0:2]
rr.columns = ['Title', 'Correlation', 'Genre']
rr = rr.drop('Genre', axis=1)
rr
```

```python
result = pd.merge(rr, df1, on='Title', how='inner')
result
```
