https://movie.naver.com/movie/board/review/list.naver?&page=1

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

uri = 'https://movie.naver.com/movie/board/review/list.naver?&page='
data = []

for page in range(1, 1001):
    target = uri+str(page)
    req  = requests.get(target)
    soup = BeautifulSoup(req.content,'html.parser')
    rows = soup.select('tbody > tr')

    for row in rows:
        title = row.select('td.movie > a')
        href = title[0]['href']
        author = row.select('td.author > a')
        score = row.select('div.mask > img')
        data.append([title[0].text.strip(), title[0]['href'], author[0].text.strip(), score[0]['alt']])
len(data)
```

```python
data
```

```python
pd_data=pd.DataFrame(data, columns=['Title','Code','Author','Score'])
pd_data
```

```python
pd_data['Code'] = pd_data['Code'].str[-6:]
pd_data
```

```python
pd_data['Code'] = pd_data['Code'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
pd_data
```

```python
pd_data['Score'] = pd_data['Score'].str.strip('점')
pd_data
```

```python
pd_data.to_excel('./saves/naver_movie_score_all.xls', index=False)
```
