```python
contents=list()

for body in bodys:
    num=body.select('td.ac.num')
    movie=body.select('a.movie')
    score=body.select('td>div>em')
    id=body.select('a.author')
    
    #print(num[0],movie[0],score[0],id[0])
    reviews=[num[0].text,movie[0].text,score[0].text,id[0].text] 
    contents.append(reviews)
    
contents
```

```python
import pandas as pd
pd_data=pd.DataFrame(contents, columns=['num','movies','score','id'])
pd_data
```
