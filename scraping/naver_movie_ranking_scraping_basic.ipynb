```python
path = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver'
req = requests.get(path)
```

```python
req.status_code
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(req.content, 'html.parser')
soup
```

```python
ranks = soup.select('img[width="14"]')
ranks, type(ranks)
```

```python
rank = ranks[0]
```

```python
rank['alt']
```

```python
rank['alt'].strip()
```

```python
ranking = list()
for rank in ranks:
#     print(int(rank['alt'].strip()))
    ranking.append(int(rank['alt'].strip()))
ranking
```

```python
ranking.insert(23, 24)
ranking
```

```python
movies = soup.select('div.tit3 > a')
movies, type(movies)
```

```python
movie = movies[0]
movie
```

```python
movie.text.strip()
```

```python
movie_name = list()
for movie in movies:
    
#     print(movie.text.strip())
    movie_name.append(movie.text.strip())
movie_name, type(movie_name)
```

```python
data = list(zip(ranking,movie_name))
data
```

```python
import pandas as pd
```

```python
pd_data = pd.DataFrame(data, columns=['Ranking','Naver Movie Ranking'])
pd_data
```

```python
pd_data.to_excel('./saves/naver_movie_ranking_basic.xls', index=False)
```
