# Load Dataset

```python
meta = pd.read_csv('/kaggle/input/the-movies-dataset/movies_metadata.csv')
meta.head()
```

필요한 변수 추출: movieId(영화 코드 -> 평점 데이터셋과 함칠 때 사용), original_title(영화 제목), original_language(언어 -> 영어 영화만 걸러냄), genres

```python
meta = meta[['id', 'original_title', 'original_language', 'genres']]
meta = meta.rename(columns={'id':'movieId'})
# df.rename() -> 컬럼 이름 변경
meta = meta[meta['original_language'] == 'en']
meta.head()
```

평가데이터 용량이 매우 크기 때문에 우선 small ver.으로 test

```python
ratings = pd.read_csv('/kaggle/input/the-movies-dataset/ratings_small.csv')
ratings = ratings[['userId', 'movieId', 'rating']]
# userId: 사용자 ID, movieId: 영화코드, rating: 평점
ratings.head()
```

```python
ratings.describe()
```

# Refine Dataset

```python
meta.movieId = pd.to_numeric(meta.movieId, errors='coerce')
# pd.to_numeric() -> 문자열(meta와 ratings 데이터셋의 movieID 변수)을 숫자 타입으로 변환
ratings.movieId = pd.to_numeric(ratings.movieId, errors='coerce')
```

```python
def parse_genres(genres_str):
    genres = json.loads(genres_str.replace('\'', '"'))
# 작은 따옴표(')를 큰 따옴표(")로 변환 
    genres_list = []
    for g in genres:
        genres_list.append(g['name'])
        # 장르만 남김

    return genres_list
    # 장르 리스트 생성

meta['genres'] = meta['genres'].apply(parse_genres)
# String 형태의 genres를 계산하기 쉽도록 배열형태로 가공
# df.apply() -> 각 행에 해당 함수를 적용

meta.head()
```

# Merge Meta and Ratings

```python
data = pd.merge(ratings, meta, on='movieId', how='inner')
# 평가테이블과 영화테이블을 inner방식으로 병합
# pd.merge() -> 두 개의 데이터프레임을 병합
data.head()
```

# Pivot Table

```python
matrix = data.pivot_table(index='userId', columns='original_title', values='rating')
# 병합한 데이터셋을 피벗테이블 형태로 가공
# df.pivot_table() -> 피벗테이블 생성
matrix.head(20)
```

# Pearson Correlation

```python
GENRE_WEIGHT = 0.1

def pearsonR(s1, s2):
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))
    # 분모가 클수록 상관관계가 떨어지고, 분자가 클수록 상관관계가 높아짐

# 추천함수 생성
def recommend(input_movie, matrix, n, similar_genre=True):
    input_genres = meta[meta['original_title'] == input_movie]['genres'].iloc(0)[0]

    result = []
    for title in matrix.columns:
        if title == input_movie:
            # 같은 영화를 추천받는 경우는 건너뜀
            continue

        # rating comparison
        cor = pearsonR(matrix[input_movie], matrix[title])
        
        # genre comparison
        if similar_genre and len(input_genres) > 0:
            temp_genres = meta[meta['original_title'] == title]['genres'].iloc(0)[0]
            # 현재 비교하고 있는 영화
            same_count = np.sum(np.isin(input_genres, temp_genres))
            # Input으로 넣은 영화와 비교
            # np.isin() -> 배열을 비교하여 똑같은 요소가 있으면 True를 반환
            cor += (GENRE_WEIGHT * same_count)
        
        if np.isnan(cor):
            continue
        else:
            result.append((title, '{:.2f}'.format(cor), temp_genres))
            
    result.sort(key=lambda r: r[1], reverse=True)
    # 리스트 정렬

    return result[:n]
```

# Prediction

```python
recommend_result = recommend('The Dark Knight', matrix, 10, similar_genre=True)
# 다크나이트 영화를 보고 비슷한 장르 10개를 추천
pd.DataFrame(recommend_result, columns = ['Title', 'Correlation', 'Genre'])
```
